```python
#Pregunta 3. Regresion lineal por minimos cuadrados

import numpy as np
import matplotlib.pyplot as plt

#imprementacion de python
T = np.array([-50, -30, 0, 60, 90, 110])
c = np.array([1270, 1280, 1350, 1480, 1580, 1700])

#calculo de la pendiente (m) y la odernada al origen (b)

n = len(T)
m = (n * np.sum(T * c) - (np.sum(T) * np.sum(c))) / (n * np.sum(T**2) - (np.sum(T))**2)
b = (np.sum(c) - m * np.sum(T)) / n

#Ecuacion de regresion lineal
regresion_line = m * T + b

#visualizacion de los resultados
plt.scatter(T, c, label='Datos')
plt.plot(T, regresion_line, color='r', label='regresion')
plt.xlabel('variable independiente T')
plt.ylabel('variable dependiente c')
plt.title('regresion lineal por minimos')
plt.show()

#ecuacion
print(f'Ecuacion de regresion: c = {m:.2f}T + {b:.2}')
```

```python
#Pregunta 3. Regresion para determinar un modelo para predecir c como funcion de T

import numpy as np
import matplotlib.pyplot as plt

#imprementacion en Python del ajuste
T = np.array([-50, -30, 0, 60, 90, 110])
c = np.array([1270, 1280, 1350, 1480, 1580, 1700])

#Ajuste lineal
coeficientes = np.polyfit(T, c, 2)
Linea_ajustada = np.polyval(coeficientes, T)

#visualizacion de los resultados
plt.scatter(T, c, label='Datos experimentales')
plt.plot(T, Linea_ajustada, label='Ajuste Lineal', color='r')
plt.xlabel('Temperatura(C)')
plt.ylabel('capacidad calorifica')
plt.legend()
plt.show()
```
